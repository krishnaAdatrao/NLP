## Prototype

```python
import numpy as np
import math

given_text = "a b c d f a b b c c d".lower().replace(',', ' ,').replace('.', ' .')

text_words = given_text.split()

probability_events = [("P({} | {})".format(text_words[i+1], text_words[i])) for i in range(len(text_words) - 1)]
occurence_event, combined_event, unique_event, prob_bigram = [], {}, {}, {}

for i in range(len(text_words) - 1): 
    
    occurence_event.append((text_words[i], text_words[i+1]))
    
    if (text_words[i], text_words[i+1]) in combined_event:
        combined_event[(text_words[i], text_words[i + 1])] += 1
    else:
        combined_event[(text_words[i], text_words[i + 1])] = 1
        
    if text_words[i] in unique_event:
        unique_event[text_words[i]] += 1
    else:
        unique_event[text_words[i]] = 1
        
for k, j in zip(probability_events, occurence_event):
    prob_bigram[k] = (combined_event.get(j))/unique_event.get(j[0])

print("\n\n", "Probability of single element text:", "\n")
print("\n", "P(Text):", math.prod(prob_bigram.values()), "\n")
prob_bigram
# training_set = "a a b c h d b f b c h h a d f f a h b b b a a h h c c h d d f a b f f h c c d f f h h h f c f c a a c c d d d".lower().replace(',', ' ,').replace('.', ' .')
# testing_set = "a b c d f a b b c c d".lower().replace(',', ' ,').replace('.', ' .')
# sentence_test = "h h f c f c a a c c d d d".lower().replace(',', ' ,').replace('.', ' .')
```

# Constructing Bi-Gram Language Model for Text

Example text: “The day was grey and bitter cold, and the dogs would not take the scent. The big black bitch had taken one sniff at the bear tracks, backed off, and skulked back to the pack with her tail between her legs.”

```python
import numpy as np
import math

given_text = """<\s> The day was grey and bitter cold, and the dogs would not take the scent. 
The big black bitch had taken one sniff at the bear tracks, backed off, 
and skulked back to the pack with her tail between her legs. <\s>""".lower().replace(',', ' ,').replace('.', ' .')

text_words = given_text.split()

probability_events = [("P({} | {})".format(text_words[i+1], text_words[i])) for i in range(len(text_words) - 1)]
occurence_event, combined_event, unique_event, prob_bigram = [], {}, {}, {}

for i in range(len(text_words) - 1): 
    
    occurence_event.append((text_words[i], text_words[i+1]))
    
    if (text_words[i], text_words[i+1]) in combined_event:
        combined_event[(text_words[i], text_words[i + 1])] += 1
    else:
        combined_event[(text_words[i], text_words[i + 1])] = 1
        
    if text_words[i] in unique_event:
        unique_event[text_words[i]] += 1
    else:
        unique_event[text_words[i]] = 1
        
for k, j in zip(probability_events, occurence_event):
    prob_bigram[k] = (combined_event.get(j))/unique_event.get(j[0])

print("\n\n", "Probability of Text:", "\n")
print("\n", "P(Text):", math.prod(prob_bigram.values()), "\n")
prob_bigram
```
